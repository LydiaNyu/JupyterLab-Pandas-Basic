Instruction 6: describe the kind of buildings that were sold

The first position of the Building Class is a letter that is used to describe a general class of properties (for example “A” signifies one-family homes, “O” signifies office buildings. “R” signifies condominiums).

Therefore, we need to classify the kind of buildings through the first letter in the 'BUILDING CLASS AT PRESENT' column.

Then through using series.str.extract method, only extract A, O, or R letter.

Then count these appearance of this letters. 

```python
classCount = (df['BUILDING CLASS AT PRESENT']).str.extract(pat = '(A?O?R?)').value_counts()
print(classCount)
```

Then we could know through above results, that which index corresponds to A, O, R, or nothing. Since there only exists 4 rows, we will manually change the index to 'condominiums','other','office buildings','one-family homes'. Then use plot method to create a visualization.

```python
classCount.index  =['condominiums','other','office buildings','one-family homes']
classCount.plot(kind="bar")
```

Instruction 8:bin the prices of properties sold

Steps: First create bins according to the summary statitics.
Then create the visualization after binning these values.
Since these use sort_index method in order to let bins appear in asecending order.

```python
df[' SALE PRICE '].describe()
```

```python
bins = [-0.1, 0.001, 3.000000e+05, 8.750000e+05 ,2.100000e+06, 2.155000e+09]
labels = ['0value','0-25%','25%-50%','50%-75%','75%-max']
binResult = pd.cut(df[' SALE PRICE '],bins, labels = labels).value_counts()
binResult = binResult.sort_index()
binResult.plot(kind="bar")
```

what is unusual about the dataset's prices?
When I try to create this visualization, I notice that there exists many property sold with price of 0 dollar, which means transfer property. It could not be generally classified as low price. So I specifically take these property with 0 value outside and make a specific bin for it.

It might affect the earlier caculations since property with price 0 will be caculated into statitics, but its real price will not be 0.

We can fix it by ignoring these property with price 0 which doing the caculation.


Instruction 9:

create a visualization the shows the relationship (if any) between 
the price sold and the number of square footage of all the areas 
of a property within a building or a structure

Steps:

First, we replace the ',' with '' in 'GROSS SQUARE FEET" column and then change it to number type.

Since there exists many property with 0 price, but 0 price does not truly reflects its value. So I specificallly ignore these property. And then these exists some missing value in 'GROSS SQUARE FEET" column, I also ignore these property.

Then I use a scatterplot to plot the the relationship (if any) between 
the price sold and the number of square footage of all the areas 
of a property within a building or a structure.

```python
df['GROSS SQUARE FEET'] = df['GROSS SQUARE FEET'].str.replace(',','')
df['GROSS SQUARE FEET'] = pd.to_numeric(df['GROSS SQUARE FEET'],errors = 'coerce')
FeetList = df[(df['GROSS SQUARE FEET'].notnull()) & (df[' SALE PRICE '] != 0) ]
FeetList
```

```python
FeetList.plot.scatter(x = ' SALE PRICE ', y = 'GROSS SQUARE FEET')
```

Instruction 10:  which month did the least amount of sales occur

Steps: we will first transform SALE DATE column to datetime64[ns] type and then count the propery sales for each month.  Then sort according to values.

SO we could get that the least amount of sales occur in August 

```python
df['SALE DATE'] = pd.to_datetime(df['SALE DATE'])
df['SALE DATE'].dt.month.value_counts().sort_values()
```

```python
countMonth = df['SALE DATE'].dt.month.value_counts().sort_index()
countMonth.plot(kind="bar")
```

the least amount of sales occur in August .
According to the above visualization, we could conclude that property sales reaches a lull in August and November.
